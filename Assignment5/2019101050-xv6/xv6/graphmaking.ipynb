```python
import csv
x = []
y = [[] for i in range(0, 100)]
with open('data2.csv', 'r') as csv_file:
    reader = csv.reader(csv_file)
    diff = 0
    for row in reader:
        x.append(int(row[0]) - diff) 
        diff = x[0]
        for i in range(1, 50):
            y[i].append(int(row[i]))
```

```python
x[0] = 0
y
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 5

plt.xlabel ('Time')
plt.ylabel ('Queue No')
for i in range(1, 50):
    plt.plot(x, y[i])
plt.legend(["CPU then IO", "Pure CPU", "Pure IO", "IO then CPU"])
plt.savefig("MLFQ-Prio.jpg")
plt.show()
```
